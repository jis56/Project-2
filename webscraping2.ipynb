```python

from bs4 import BeautifulSoup
import requests
import os
from splinter import Browser
import time
import pandas as pd
import inspect
```

```python
#scraping foundation page soup for list of url

page = requests.get('https://www.ulta.com/makeup-face-foundation?N=26y5')
soup = BeautifulSoup(page.content, 'html.parser')
time.sleep(1)
foundationurl_list = []

for a in soup.find_all('a', class_='product'):
    try:
        foundationurl_list.append(a['href'])
    except KeyError:
        pass
    
print(foundationurl_list)
```

```python
brand_list = []
product_list = []
price_list = []
rating_list = []
img_list = []
category_list = []

executable_path = {'executable_path': r"C:\Chromedrive\chromedriver.exe"}
browser = Browser('chrome', **executable_path, headless=False)

for url in foundationurl_list:
    
    if url.startswith(r'/'):
        try: 
            browser.visit('https://www.ulta.com'+ url)
            time.sleep(3)

            html = browser.html
            soup = BeautifulSoup(html, 'html.parser')

            brand = soup.find_all('p', class_='Text--$magenta-50')[0].text
            brand_list.append(brand)

            product = soup.find_all('span', class_='Text--text-20')[0].text
            product_list.append(product)

            price = soup.find_all('span', class_='Text--title-6')[0].text
            price_list.append(price)

            rating = soup.find_all('div', class_='RatingPanel__panel--foreground')[0].text
            rating_list.append(rating)

            img = soup.find('div', class_= 's7staticimage').find('img')['src']
            img_list.append(img)

            category = soup.find_all('a', class_='Anchor Tertiary')[3].text
            category_list.append(category)
            
        except:
            pass
        
    else:
        
        try:
            browser.visit(url)
            time.sleep(3)

            html = browser.html
            soup = BeautifulSoup(html, 'html.parser')

            brand = soup.find_all('p', class_='Text--$magenta-50')[0].text
            brand_list.append(brand)

            product = soup.find_all('span', class_='Text--text-20')[0].text
            product_list.append(product)

            price = soup.find_all('span', class_='Text--title-6')[0].text
            price_list.append(price)

            rating = soup.find_all('div', class_='RatingPanel__panel--foreground')[0].text
            rating_list.append(rating)

            img = soup.find('div', class_= 's7staticimage').find('img')['src']
            img_list.append(img)

            category = soup.find_all('a', class_='Anchor Tertiary')[3].text
            category_list.append(category)
        
        except:
            pass
```
